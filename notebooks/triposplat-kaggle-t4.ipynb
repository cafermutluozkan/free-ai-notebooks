### 🚀 What is TripoSplat?

Converts a **single 2D image** into **3D Gaussian Splats** — viewable in any 3DGS viewer. Up to **262,144 gaussians** for high-quality results.

| Feature | Detail |
|---|---|
| **Model** | TripoSplat (3.78 GB) |
| **GPU** | T4 (free) |
| **Dependencies** | Only 5 packages (no transformers/diffusers!) |
| **Output** | .ply / .splat files + in-browser 3D viewer |
| **License** | MIT |
