## 🖥️ Local Setup Guide

### ✅ Requirements
| Requirement | Minimum | Recommended |
|---|---|---|
| **GPU** | NVIDIA 12 GB VRAM | 16 – 24 GB VRAM |
| **RAM** | 16 GB | 32 GB |
| **Disk** | 25 GB free | 40 GB free |
| **Python** | 3.10 | 3.10 – 3.12 |
| **CUDA** | 12.1 | 12.4 |

### ⚡ Quick Start (one-time setup)
1. Install **Python 3.10 – 3.12** → [python.org/downloads](https://www.python.org/downloads/)
2. Install **CUDA 12.x + PyTorch** → [pytorch.org/get-started/locally](https://pytorch.org/get-started/locally/)
3. Install **Jupyter**: open a terminal and run `pip install jupyter`
4. Open this notebook and **run cells 1 → 2 → 3** — everything else is fully automatic!

> 💡 **No GPU?** The notebook will warn you. CPU-only mode works but generation takes 30+ minutes per video.
> 
> 🔁 **Already set up?** Skip Cell 1 and go straight to Cell 2 or Cell 3.

```python
# ╔══════════════════════════════════════════════════════════════════╗
# ║  Cell 1 — System Check                                          ║
# ║  Run this first to verify your environment is ready.            ║
# ╚══════════════════════════════════════════════════════════════════╝
import sys, platform, shutil

print('=' * 55)
print('  LTX-Video Generator — System Check')
print('=' * 55)

# ── Python ────────────────────────────────────────────────────────
py_ver = sys.version_info
py_ok  = (3, 10) <= (py_ver.major, py_ver.minor) <= (3, 12)
print(f'\nPython  {sys.version.split()[0]}  {"✅" if py_ok else "❌  Need 3.10 – 3.12"}')

# ── PyTorch + CUDA ────────────────────────────────────────────────
try:
    import torch
    cuda_ok = torch.cuda.is_available()
    print(f'PyTorch {torch.__version__}  ✅')
    if cuda_ok:
        gpu      = torch.cuda.get_device_name(0)
        vram_gb  = torch.cuda.get_device_properties(0).total_memory / 1024**3
        vram_ok  = vram_gb >= 12
        print(f'CUDA    {torch.version.cuda}  ✅')
        print(f'GPU     {gpu}')
        print(f'VRAM    {vram_gb:.1f} GB  {"✅" if vram_ok else "⚠️  12 GB minimum recommended"}')
        if vram_gb < 12:
            print('        Generation may be slow or fail — consider a larger GPU.')
    else:
        print('CUDA    ❌  Not available — CPU-only mode (very slow!)')
        print('        Install CUDA 12.x + PyTorch GPU build from pytorch.org')
except ImportError:
    print('PyTorch ❌  Not installed!')
    print('        Run: pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu124')

# ── Disk ──────────────────────────────────────────────────────────
import os
HOME      = os.path.expanduser('~')
free_gb   = shutil.disk_usage(HOME).free / 1024**3
disk_ok   = free_gb >= 25
print(f'\nDisk    {free_gb:.1f} GB free  {"✅" if disk_ok else "⚠️  25 GB needed for models"}')

# ── OS ────────────────────────────────────────────────────────────
print(f'OS      {platform.system()} {platform.release()}')

# ── git ───────────────────────────────────────────────────────────
git_ok = shutil.which('git') is not None
print(f'git     {"✅" if git_ok else "❌  Install from git-scm.com"}')

print('\n' + '=' * 55)
if py_ok and git_ok:
    print('  ✅  System looks good — proceed to Cell 2!')
else:
    print('  ❌  Fix the issues above before continuing.')
print('=' * 55)
```

```python
# ╔══════════════════════════════════════════════════════════════════╗
# ║  Cell 2 — Install & Download  (run once, ~20 min first time)    ║
# ║  Skip this cell if you have already run it before.              ║
# ╚══════════════════════════════════════════════════════════════════╝
import os, sys, subprocess
from pathlib import Path

HOME      = os.path.expanduser('~')
WAN2GP    = os.path.join(HOME, 'Wan2GP')
MODEL_DIR = os.path.join(WAN2GP, 'models')

# ── 1 / 4  Install base packages ─────────────────────────────────
print('[1/4] Installing dependencies...')
pkgs = ['huggingface_hub', 'gradio>=4.0', 'gguf', 'soundfile', 'mmgp']
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', '--upgrade'] + pkgs, check=True)

import torch
print(f'  PyTorch {torch.__version__} | CUDA available: {torch.cuda.is_available()}')

# ── 2 / 4  Clone Wan2GP ──────────────────────────────────────────
print('[2/4] Setting up Wan2GP...')
if not os.path.exists(WAN2GP):
    subprocess.run(['git', 'clone', '-q',
                    'https://github.com/deepbeepmeep/Wan2GP.git', WAN2GP], check=True)
    print('  Cloned Wan2GP')
else:
    print('  Wan2GP already exists — pulling latest...')
    subprocess.run(['git', '-C', WAN2GP, 'pull', '-q'], check=False)

subprocess.run([sys.executable, '-m', 'pip', 'install', '-q',
                '-r', os.path.join(WAN2GP, 'requirements.txt')], check=True)
print('  Requirements installed')

# ── 3 / 4  Download models ───────────────────────────────────────
print('[3/4] Downloading models (~17.8 GB total)...')
print('  This takes 10–30 min depending on your connection.  Already downloaded files are skipped.')
from huggingface_hub import hf_hub_download

def dl(repo, filename, dest_dir, dest_name=None):
    Path(dest_dir).mkdir(parents=True, exist_ok=True)
    target = dest_name or os.path.basename(filename)
    fp = os.path.join(dest_dir, target)
    if os.path.exists(fp):
        print(f'  ✓ {target}  (already downloaded)')
        return
    print(f'  ↓ Downloading {target}...')
    hf_hub_download(repo_id=repo, filename=filename,
                    local_dir=dest_dir, local_dir_use_symlinks=False)
    # hf_hub_download preserves subfolder structure; flatten to dest_dir
    tmp = os.path.join(dest_dir, filename)
    if os.path.exists(tmp) and tmp != fp:
        os.makedirs(os.path.dirname(fp), exist_ok=True)
        os.rename(tmp, fp)
    print(f'  ✅ {target}')

# Transformer (GGUF)
dl('Abiray/LTX-2.3-22B-DISTILLED-1.1-GGUF',
   'LTX-2.3-22B-distilled-1.1-Q4_K_M.gguf', MODEL_DIR,
   'ltx-2.3-22b-distilled-1.1-Q4_K_M.gguf')

# Companion weights
for _f in [
    'ltx-2.3-22b-distilled-lora-384.safetensors',
    'ltx-2.3-22b_embeddings_connector.safetensors',
    'ltx-2.3-22b_text_embedding_projection.safetensors',
    'ltx-2.3-22b_vae.safetensors',
    'ltx-2.3-22b_audio_vae.safetensors',
    'ltx-2.3-22b_vocoder.safetensors',
    'ltx-2.3-spatial-upscaler-x2-1.1.safetensors',
    'ltx-2.3-temporal-upscaler-x2-1.0.safetensors',
]:
    dl('DeepBeepMeep/LTX-2', _f, MODEL_DIR)

# Gemma text encoder
GEMMA     = 'gemma-3-12b-it-qat-q4_0-unquantized'
GEMMA_DIR = os.path.join(MODEL_DIR, GEMMA)
for _gf in [
    'gemma-3-12b-it-qat-q4_0-unquantized_quanto_bf16_int8.safetensors',
    'added_tokens.json', 'chat_template.json', 'config_light.json',
    'generation_config.json', 'preprocessor_config.json',
    'processor_config.json', 'special_tokens_map.json',
    'tokenizer.json', 'tokenizer.model', 'tokenizer_config.json',
]:
    dl('DeepBeepMeep/LTX-2', f'{GEMMA}/{_gf}', GEMMA_DIR, dest_name=_gf)

# ── 4 / 4  Verify ────────────────────────────────────────────────
print('[4/4] Verifying downloads...')
import glob

transformer_path = os.path.join(MODEL_DIR, 'ltx-2.3-22b-distilled-1.1-Q4_K_M.gguf')
assert os.path.exists(transformer_path), f'Missing: {transformer_path}'

hits = (glob.glob(f'{GEMMA_DIR}/*.safetensors') +
        glob.glob(f'{GEMMA_DIR}/**/*.safetensors', recursive=True))
quanto = [f for f in hits if 'quanto' in f]
assert quanto, f'Missing Gemma safetensors in {GEMMA_DIR}'
text_encoder_file = quanto[0]

print(f'  ✅ transformer      : {os.path.basename(transformer_path)}')
print(f'  ✅ text encoder     : {os.path.basename(text_encoder_file)}')
print()
print('=' * 55)
print('  ✅  Setup complete — run Cell 3 to launch!')
print('=' * 55)
```

```python
# ╔══════════════════════════════════════════════════════════════════╗
# ║  Cell 3 — Launch  (run every time you want to generate videos)  ║
# ║  A browser tab will open automatically when ready.              ║
# ╚══════════════════════════════════════════════════════════════════╝
import os, sys, gc, glob, random, traceback, webbrowser
from pathlib import Path
from datetime import datetime

HOME      = os.path.expanduser('~')
WAN2GP    = os.path.join(HOME, 'Wan2GP')
MODEL_DIR = os.path.join(WAN2GP, 'models')
GEMMA     = 'gemma-3-12b-it-qat-q4_0-unquantized'
GEMMA_DIR = os.path.join(MODEL_DIR, GEMMA)
OUTPUT_DIR = Path(os.path.join(HOME, 'ltx-videos'))
OUTPUT_DIR.mkdir(exist_ok=True)

os.chdir(WAN2GP)
if WAN2GP not in sys.path:
    sys.path.insert(0, WAN2GP)

os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
os.environ['TOKENIZERS_PARALLELISM']  = 'false'

import torch
import gradio as gr

device    = 'cuda' if torch.cuda.is_available() else 'cpu'
dtype     = torch.float16 if device == 'cuda' else torch.float32
vram_gb   = torch.cuda.get_device_properties(0).total_memory / 1024**3 if device == 'cuda' else 0

print(f'Device : {device.upper()}', end='')
if device == 'cuda':
    print(f'  |  {torch.cuda.get_device_name(0)}  |  {vram_gb:.1f} GB VRAM')
else:
    print('  ⚠️  CPU mode — generation will be very slow!')

# ── Resolve model paths ───────────────────────────────────────────
transformer_path = os.path.join(MODEL_DIR, 'ltx-2.3-22b-distilled-1.1-Q4_K_M.gguf')
if not os.path.exists(transformer_path):
    raise FileNotFoundError(
        f'Transformer not found: {transformer_path}\n'
        'Please run Cell 2 first to download the models.')

hits   = (glob.glob(f'{GEMMA_DIR}/*.safetensors') +
          glob.glob(f'{GEMMA_DIR}/**/*.safetensors', recursive=True))
quanto = [f for f in hits if 'quanto' in f]
if not quanto:
    raise FileNotFoundError(
        f'Gemma text encoder not found in {GEMMA_DIR}\n'
        'Please run Cell 2 first.')
text_encoder_file = quanto[0]

print(f'Transformer  : {os.path.basename(transformer_path)}')
print(f'Text encoder : {os.path.basename(text_encoder_file)}')

# ── Load Wan2GP ───────────────────────────────────────────────────
print('\nLoading model (this takes 1–3 min)...')

import shared.qtypes.gguf
import models.ltx2.ltx2 as ltx2_mod

def _patched_load(path, fallback_config_path=None):
    from mmgp import quant_router
    if isinstance(path, (list, tuple)): path = path[0] if path else ''
    if not path: return {}
    try:
        _, meta = quant_router.load_metadata_state_dict(path)
        if meta:
            raw = meta.get('config')
            if raw:
                cfg = ltx2_mod._normalize_config(raw)
                if cfg: return cfg
    except Exception:
        pass
    return {}
ltx2_mod._load_config_from_checkpoint = _patched_load

from mmgp import offload
from shared.utils import files_locator as fl
fl.set_checkpoints_paths(['models', 'ckpts', '.'])
from models.ltx2.ltx2_handler import family_handler

base_model_type = 'ltx2_22B'
model_def = {'ltx2_pipeline': 'distilled'}
model_def.update(family_handler.query_model_def(base_model_type, model_def))

ltx2_model, pipe = family_handler.load_model(
    model_filename=transformer_path,
    model_type='ltx2_22B_distilled',
    base_model_type=base_model_type,
    model_def=model_def,
    dtype=dtype,
    VAE_dtype=dtype,
    text_encoder_filename=text_encoder_file,
)

# ── mmgp memory profile ───────────────────────────────────────────
# Profile 4 = aggressive CPU offload; works on 12 GB+ VRAM
# Increase transformer budget if you have 24+ GB VRAM for faster speed
offload.profile(
    pipe, profile_no=4, quantizeTransformer=False,
    convertWeightsFloatTo=dtype,
    budgets={
        'transformer': 6000, 'text_encoder': 1500,
        'video_encoder': 2000, 'video_decoder': 3000,
        'audio_encoder': 1000, 'audio_decoder': 1000,
        'vocoder': 500,  'spatial_upsampler': 1500,
        'vae': 1000,     '*': 1000,
    },
)
offload.shared_state['_attention'] = 'sdpa'
print('✅ Model loaded — starting Gradio...')

# ── Helpers ───────────────────────────────────────────────────────
def get_res(base_res, aspect):
    bases  = {'1080p': 1088, '720p': 704, '540p': 544, '480p': 480}
    ratios = {'16:9': 16/9, '4:3': 4/3, '1:1': 1.0, '3:4': 3/4, '9:16': 9/16}
    b, r = bases.get(base_res, 704), ratios.get(aspect, 16/9)
    h, w = (b, int(b*r)) if r >= 1.0 else (int(b/r), b)
    return (w//32)*32, (h//32)*32

# ── Generate ──────────────────────────────────────────────────────
@torch.inference_mode()
def generate(prompt, img_start, img_end, seed, duration,
             resolution, aspect_ratio, guide_scale, num_steps,
             progress=gr.Progress()):
    try:
        gc.collect()
        if device == 'cuda': torch.cuda.empty_cache()
        progress(0, desc='Starting...')

        dur_map    = {'2s (49f)': 49, '3s (73f)': 73, '5s (121f)': 121,
                      '8s (193f)': 193, '10s (241f)': 241}
        num_frames = dur_map.get(duration, 121)
        width, height = get_res(resolution, aspect_ratio)
        seed = int(seed) if (seed is not None and seed >= 0) else random.randint(0, 2**32-1)

        from PIL import Image
        img_s = Image.open(img_start).convert('RGB') if img_start else None
        img_e = Image.open(img_end).convert('RGB')   if img_end   else None

        total_steps = [8]; current_step = [0]; current_pass = [1]
        def cb(step, latent, is_start, override_num_inference_steps=None,
               pass_no=None, **kwargs):
            if is_start:
                if override_num_inference_steps: total_steps[0] = override_num_inference_steps
                if pass_no: current_pass[0] = pass_no
                current_step[0] = 0; return
            current_step[0] += 1
            frac = current_step[0] / max(total_steps[0], 1)
            frac = (0.73 + 0.22*frac) if current_pass[0]==2 else (frac*0.73)
            progress(min(frac, 0.95),
                     desc=f'Stage {current_pass[0]}: {current_step[0]}/{total_steps[0]}')

        result = ltx2_model.generate(
            input_prompt=prompt, image_start=img_s, image_end=img_e,
            height=height, width=width, frame_num=num_frames,
            fps=24.0, seed=seed, callback=cb,
            VAE_tile_size=256, guide_scale=float(guide_scale),
            sampling_steps=int(num_steps), guide_phases=2,
            n_prompt='', enhance_prompt=False,
            input_video_strength=1.0,
        )

        progress(0.97, desc='Saving...')
        # Unpack result (dict / tuple / raw tensor)
        if isinstance(result, dict):
            for _k in ('video', 'frames', 'samples', 'output', 'images'):
                if _k in result and result[_k] is not None:
                    video_tensor = result[_k]; break
            else:
                video_tensor = next(iter(result.values()))
        elif isinstance(result, (tuple, list)):
            video_tensor = result[0]
        else:
            video_tensor = result

        if video_tensor is None:
            return None, 'Error: No video produced'

        from shared.utils.audio_video import save_video
        video_tensor   = video_tensor.cpu()
        ts             = datetime.now().strftime('%Y%m%d_%H%M%S')
        out            = str(OUTPUT_DIR / f'video_{ts}.mp4')
        video_for_save = video_tensor.unsqueeze(0).float() / 127.5 - 1.0
        save_video(tensor=video_for_save, save_file=out,
                   fps=24.0, normalize=True, value_range=(-1, 1))

        del video_tensor, video_for_save
        gc.collect()
        if device == 'cuda': torch.cuda.empty_cache()
        progress(1.0, desc='Done!')
        return out, f'✅ Seed: {seed} | {width}x{height} | {num_frames}f | saved to {out}'

    except Exception as e:
        traceback.print_exc()
        gc.collect()
        if device == 'cuda': torch.cuda.empty_cache()
        return None, f'Error: {e}'

# ── Gradio UI ─────────────────────────────────────────────────────
with gr.Blocks(title='LTX-Video Generator — Local', theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        '# 🎬 LTX-Video Generator &nbsp; <small>Local Edition</small>\n'
        f'**{torch.cuda.get_device_name(0) if device=="cuda" else "CPU mode"}** '
        f'| {vram_gb:.1f} GB VRAM | Wan2GP + mmgp | Q4_K_M GGUF'
    )

    with gr.Row():
        with gr.Column(scale=3):
            prompt = gr.Textbox(label='Prompt', lines=4,
                                placeholder='A cinematic drone shot of a mountain lake at golden hour...')
            with gr.Accordion('🖼️ Image-to-Video (optional — attach start/end frames)', open=False):
                with gr.Row():
                    img_start = gr.Image(type='filepath', label='Start Frame')
                    img_end   = gr.Image(type='filepath', label='End Frame')

        with gr.Column(scale=1):
            seed         = gr.Number(-1, label='Seed  (-1 = random)', precision=0)
            duration     = gr.Dropdown(
                ['2s (49f)', '3s (73f)', '5s (121f)', '8s (193f)', '10s (241f)'],
                value='5s (121f)', label='Duration')
            resolution   = gr.Dropdown(['1080p','720p','540p','480p'],
                                        value='720p', label='Resolution')
            aspect_ratio = gr.Dropdown(['16:9','4:3','1:1','3:4','9:16'],
                                        value='16:9', label='Aspect Ratio')
            guide_scale  = gr.Slider(1.0, 8.0, 3.0, step=0.5, label='Guidance Scale')
            num_steps    = gr.Slider(2, 8, 8, step=1, label='Steps')

    btn       = gr.Button('🎬 Generate Video', variant='primary', size='lg')
    video_out = gr.Video(label='Output Video')
    status    = gr.Textbox(label='Status', interactive=False)

    btn.click(
        generate,
        [prompt, img_start, img_end, seed, duration,
         resolution, aspect_ratio, guide_scale, num_steps],
        [video_out, status]
    )

    gr.Markdown(
        f'> 💾 Videos saved to **`{OUTPUT_DIR}`**\n\n'
        '> 🔔 Like this notebook? '
        '[Subscribe on YouTube](https://www.youtube.com/@thebuildai?sub_confirmation=1) '
        'and [Star on GitHub](https://github.com/cafermutluozkan/free-ai-notebooks)!'
    )

# Open browser automatically after a short delay
import threading, time
def _open_browser():
    time.sleep(2)
    webbrowser.open('http://127.0.0.1:7860')
threading.Thread(target=_open_browser, daemon=True).start()

demo.queue(max_size=2)
demo.launch(
    server_name='127.0.0.1',
    server_port=7860,
    share=False,
    allowed_paths=[str(OUTPUT_DIR)],
    inbrowser=False,   # we handle browser open manually above
)
```
